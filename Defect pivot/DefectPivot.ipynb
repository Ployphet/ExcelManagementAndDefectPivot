```python
import pandas as pd

linerAirScrap_path = r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\Defect actual\LinerAirScrap_Pivot.csv"
linerAirRework_path = r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\Defect actual\LinerAirRework_Pivot.csv"
curing_actual_path = r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\Curing actual\CuringActual.csv"

df_actual = pd.read_csv(curing_actual_path)

dfScrap_defect = pd.read_csv(linerAirScrap_path)
mergedScrap_df = pd.merge(
    df_actual,
    dfScrap_defect,
    on=["Date", "Size"],
    how="left"
)
mergedScrap_df["Defect Count"] = mergedScrap_df["Defect Count"].fillna(0)
outputScrap_path = r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\LinerAirScrapPivot.csv"
mergedScrap_df.to_csv(outputScrap_path, index=False)

dfRework_defect = pd.read_csv(linerAirRework_path)
mergedRework_df = pd.merge(
    df_actual,
    dfRework_defect,
    on=["Date", "Size"],
    how="left"
)
mergedRework_df["Defect Count"] = mergedRework_df["Defect Count"].fillna(0)
outputRework_path = r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\LinerAirReworkPivot.csv"
mergedRework_df.to_csv(outputRework_path, index=False)
```
