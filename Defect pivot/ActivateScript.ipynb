```python
import tkinter as tk
import subprocess
from tkinter import messagebox

def run_notebook(script_path):
    jupyter_path = r"C:\Users\Ployphet.veerasak\AppData\Local\anaconda3\Scripts\jupyter.exe"

    try:
        result = subprocess.run(
            [jupyter_path, "nbconvert", "--execute", "--to", "notebook", script_path],
            check=True, capture_output=True, text=True
        )
        return True, ""
    except subprocess.CalledProcessError as e:
        return False, f"Error: {e}\n{e.output}"

def run_allscripts():
    scripts = [
        r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\Defect actual\LinerAir.ipynb",
        r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\Curing actual\CuringActual.ipynb",
        r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\DefectPivot.ipynb"
    ]

    errors = []
    for script in scripts:
        success, message = run_notebook(script)
        if not success:
            errors.append((script, message))

    if errors:
        with open("log.txt", "w") as log_file:
            for script, error in errors:
                log_file.write(f"Failed: {script}\n{error}\n\n")
        messagebox.showerror("Error", "Some scripts failed! Please, check log.txt for details.")
    else:
        messagebox.showinfo("Success", "Ran scripts successfully!")

window = tk.Tk()
window.title("Run Defect and Curing Actual Script")
window.geometry("300x150")
window.iconbitmap(r"C:\Users\Ployphet.veerasak\Desktop\Defect\Liner air\Image\BridgestoneICO.ico")

button = tk.Button(window, text="Run Scripts", command=run_allscripts)
button.pack(pady=20)

window.mainloop()
```
